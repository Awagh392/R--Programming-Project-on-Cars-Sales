```r
# Loading libraries
library(tidyverse)
library(lubridate)
library(scales)
library(ggplot2)
library(dplyr)
library(readr)
library(GGally)
```

```r
# Reading dataset
car_sales_data <- read.csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/car%20data.csv")

```

**Visualisation**

```r
ggplot(car_sales_data, aes(x=Year, y=Selling_Price)) + 
  geom_point(color="blue") +
  ggtitle("Scatter plot of Year vs. Selling Price") +
  xlab("Year") +
  ylab("Selling Price")
```

```r
ggplot(car_sales_data, aes(x=Fuel_Type)) +
  geom_bar(fill="blue") +
  ggtitle("Number of Cars Sold by Fuel Type") +
  xlab("Fuel Type") +
  ylab("Number of Cars")
```

```r
ggplot(car_sales_data, aes(x=Fuel_Type, y=Selling_Price)) +
  geom_boxplot(fill="blue") +
  ggtitle("Price by Fuel Type") +
  xlab("Fuel Type") +
  ylab("Price")
```

```r
ggplot(car_sales_data, aes(x="", fill=Transmission)) +
  geom_bar(width = 1) +
  coord_polar("y", start=0) +
  ggtitle("Number of Cars Sold by Transmission Type") +
  scale_fill_manual(values=c("blue", "red")) +
  labs(fill="Transmission") +
  theme_void()
```

```r
ggplot(car_sales_data, aes(x=Year, y=Selling_Price)) +
  geom_line(color="blue") +
  ggtitle("Car Prices over Time") +
  xlab("Year") +
  ylab("Price")
```

```r
library(GGally)
ggpairs(car_sales_data[, c("Kms_Driven", "Fuel_Type", "Present_Price")], 
        lower = list(continuous = wrap("points", size = 0.5, alpha = 0.5)), 
        diag = list(continuous = wrap("densityDiag", alpha = 0.5))) +
  ggtitle("Scatter Plot Matrix of Kms_Driven , Fule_Type, Present_Price")
```
